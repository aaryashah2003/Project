```python
from PIL import Image 
  
# Opens an image in RGB mode 
im = Image.open('/content/drive/MyDrive/Internship2/Day07/try4.jpg')

  
# Setting the points for cropped image 
left = 512
top = 254
right = 1133
bottom = 465

im1 = im.crop((left, top, right, bottom)) 
im1 = im.rotate(-9).crop((left, top, right, bottom))
# Shows the image in image viewer 
im1.save("/content/drive/MyDrive/Internship2/Day07/temp/car4_segmented.jpg")
Image.open("/content/drive/MyDrive/Internship2/Day07/temp/car4_segmented.jpg")
```

```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
img = Image.open("/content/drive/MyDrive/Internship2/Day07/temp/car4_segmented.jpg")

def ocr_core(img):
    text = pytesseract.image_to_string(Image.open(img)) 
    return text

print(ocr_core("/content/drive/MyDrive/Internship2/Day07/temp/car4_segmented.jpg"))
```
