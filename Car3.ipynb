```python
from PIL import Image 
  
# Opens an image in RGB mode 
im = Image.open('/content/drive/MyDrive/Internship2/Day07/try3.jpg')

  
# Setting the points for cropped image 
left = 330
top = 300
right = 570
bottom = 370
  
# Cropped image of above dimension 
# (It will not change orginal image) 
im1 = im.crop((left, top, right, bottom)) 
  
# Shows the image in image viewer 
im1.save("/content/drive/MyDrive/Internship2/Day07/temp/car3_segmented.jpg")
Image.open("/content/drive/MyDrive/Internship2/Day07/temp/car3_segmented.jpg")
```

```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
img = Image.open("/content/drive/MyDrive/Internship2/Day07/temp/car3_segmented.jpg")

def ocr_core(img):
    text = pytesseract.image_to_string(Image.open(img)) 
    return text

print(ocr_core("/content/drive/MyDrive/Internship2/Day07/temp/car3_segmented.jpg"))
```
